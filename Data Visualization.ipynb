```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#in order to get rid of plt.show()
```

```python
%matplotlib inline 
```

```python
x = np.arange(0, 10)
print(x)
```

```python
y = x**2
```

```python
#plotting in different styles
```

```python
plt.plot(x, y, '*')
plt.plot(x, y, 'r')
plt.plot(x, y, 'r--')
```

```python
### defining the limits of Y and X
```

```python
plt.plot(x, y)
plt.xlim(0,5)
plt.ylim(0,10)
plt.title("Random Variable")
plt.xlabel("Random X")
plt.ylabel("Random Y")
```

```python
# Showing an image in 2-D array
```

```python
matrix = np.arange(0,100).reshape(10,10)
```

```python
print(matrix)
```

### We can use cmap function to show the pixel color


```python
plt.imshow(matrix, cmap='RdYlGn') #Check cmap documentation for more styles
```

### Color map on random numbers 

```python
random_matrix = np.random.randint(low = 0,high= 1000, size= (10,10))
```

```python
plt.imshow(random_matrix, cmap='coolwarm')
plt.colorbar() #to get the color bar as well
```
